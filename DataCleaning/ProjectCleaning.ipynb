```python
import pandas as pd
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

ds = pd.read_csv('orders.csv', delimiter='|', header=None, encoding='ISO-8859-1')
ds.columns = ["Customer Code", "DoB", "Gender", "City", "Order Nr", "Order Date", "Product Nr", "Sub Category", "Category", 
            "Amount", "Price", "Expected Delivery Time", "Actual Delivery Time", "Reason of Return", "Rating", "Nr"]
       


ds.at[0,'Customer Code'] = 1019099
ds = ds.dropna(subset=['Actual Delivery Time'])
ds = ds.replace(to_replace = [ r'\\'], value = ['remove'])
ds = ds[ds['Sub Category'] != 'remove']
ds = ds[ds['Actual Delivery Time'] != -1]
ds = ds.drop(columns=["Customer Code", "DoB", "Gender", "Order Nr", "Order Date", "Amount", "Price", "Expected Delivery Time", "Reason of Return", "Rating", "Nr"])

longlat = pd.read_excel('LongLat.xlsx')
longlat.columns = ['City', 'Latitude', 'Longitude']
completeDS = pd.merge(longlat, ds, how='right', on='City')
completeDS = completeDS.dropna(subset=['Latitude', 'Longitude'])
finaldf = completeDS.sample(n = 20005)
finaldf.head(10)
```

```python
finaldf.info()
```

```python
finaldf["Product Nr"] = pd.to_numeric(finaldf["Product Nr"])
finaldf.info()
```

```python
def normalize(dataset):
    dataNorm=((dataset-dataset.min())/(dataset.max()-dataset.min()))
    dataNorm["Product Nr"]=dataset["Product Nr"]
    return dataNorm
```

```python
finaldf = finaldf.drop(columns=['City', 'Sub Category', 'Category'])
finaldf = normalize(finaldf)
finaldf.head(10)
```
